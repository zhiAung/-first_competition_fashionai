```python
import csv
import os
import numpy as np
import random

def getDataPath(lablePath, actionType) :
	pathList = []
	labelList = []
	with open(lablePath, "r") as csvFile :
		reader = csv.reader(csvFile)
		for line in reader :
			if line[1] == actionType :
				pathList.append(line[0])
				labelList.append(line[2])
	#print(pathList)
	#print(labelList)
	return pathList, labelList

def movedata(big_filepath,lebelPath):
	'''
	对文件夹中的图片进行细分
	'''
	if os.listdir(big_filepath):
		print("big_filepath存在")
	else:
		print("big_filepath不存在")
		return
	file_name_list=os.listdir(big_filepath)
	print("共有%d个文件夹需要分类"%len(file_name_list))
	
	for file_name in file_name_list:
		actionType=file_name
		pathList,labelList=getDataPath(lablePath,actionType)
		classNum=len(labelList[0])
		print("%s 共有 %d 张图片，共分 %d类"%(actionType,len(labelList),classNum))
		for i in range(classNum):
			os.mkdir(os.path.join(big_filepath,actionType,str(i)))
		print("新建文件夹完成...")

		for i,label in enumerate(labelList):
			labelList[i]=label.find("y")
		print("修改标签完成...")

		for i,path in enumerate(pathList):
			old_name=os.path.join(big_filepath,file_name,path.split("/")[-1])
			new_name=os.path.join(big_filepath,file_name,str(labelList[i]),path.split("/")[-1])
			try:
				os.rename(old_name,new_name)
			except:
				#print("%s移动失败了"%(old_name))
				pass
		print(file_name,"数据分类完成了...")

	print("数据分类完成了")

def get_validate(validate_path,train_path,rate=0.1):
	'''
	根据训练集目录生生成验证集
	rate是验证集比例
	'''
	big_class_list=os.listdir(train_path)
	print("共有%d个大类"%(len(big_class_list)))

	for big_class in big_class_list:
		if not os.path.exists(os.path.join(validate_path,big_class)):
			os.mkdir(os.path.join(validate_path,big_class))
		class_list=os.listdir(os.path.join(train_path,big_class))
		print("%s 共有 %d 个小类"%(big_class,len(class_list)))

		for small_class in class_list:
			if not os.path.exists(os.path.join(validate_path,big_class,small_class)):
				os.mkdir(os.path.join(validate_path,big_class,small_class))
			img_list=os.listdir(os.path.join(train_path,big_class,small_class))
			print("%s 小类共有 %d张图片"%(small_class,len(img_list)))

			#移动的图片数目
			move_num=int(rate*len(img_list))
			if move_num<1:
				continue
			#随机生成移动的图片的序号
			#move_array=np.random.randint(low=0,high=len(img_list),size=(move_num))
			move_array=random.sample(range(0,len(img_list)),move_num)
			#移动图片
			for num in move_array:
				old_name=os.path.join(train_path,big_class,small_class,img_list[num])
				new_name=os.path.join(validate_path,big_class,small_class,img_list[num])
				try:
					os.rename(old_name,new_name)
				except:
					pass
		print("%s 大类移动完成了"%(big_class))

def remove_validate(train_path,validate_path):
	'''
	解散验证集
	'''
	big_class_list=os.listdir(validate_path)

	for big_class in big_class_list:
		small_class_list=os.listdir(os.path.join(validate_path,big_class))
		for small_class in small_class_list:
			img_list=os.listdir(os.path.join(validate_path,big_class,small_class))
			for img in img_list:
				old_name=os.path.join(validate_path,big_class,small_class,img)
				new_name=os.path.join(train_path,big_class,small_class,img)
				os.rename(old_name,new_name)

if __name__=="__main__":
	'''
	big_filepath="G:/competition/tianchi/ali_data/fashionAI_attributes_train_20180222/base/Images/"
	lablePath="G:/label.csv"
	movedata(big_filepath,lablePath)
	'''
	
	
	train_path="/home/user01/envoriments/data/train"
	validate_path="/home/user01/envoriments/data/validate"
	get_validate(validate_path,train_path,rate=0.1)
	
	
	'''
	train_path="G:/competition/tianchi/ali_data/fashionAI_attributes_train_20180222/base/Images/"
	validate_path="G:/competition/tianchi/ali_data/fashionAI_attributes_validate_20180222/Images/"
	remove_validate(train_path,validate_path)
	'''


```
